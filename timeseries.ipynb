```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm 
```

```python
passenger = pd.read_csv("passengers.csv")
passenger.head(3)
```

```python
passenger["date"] = pd.to_datetime(passenger["date"])
passenger.set_index("date").plot()
```

```python
decomposition = sm.tsa.seasonal_decompose(passenger["passengers"], model="additive",period=12)

plt.figure(figsize=(10,8))

plt.subplot(4,1,1)
plt.plot(passenger["passengers"])
plt.title("Original Time Series")

plt.subplot(4,1,2)
plt.plot(decomposition.trend)
plt.title("Trend")

plt.subplot(4,1,3)
plt.plot(decomposition.seasonal)
plt.title("Seasonality")

plt.subplot(4,1,4)
plt.plot(decomposition.resid)
plt.title("Noise")


plt.tight_layout()  # Adjust layout to prevent overlapping titles
plt.show()
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Assuming 'series' is your DataFrame and 'passengers' is a column in it
decomposition = sm.tsa.seasonal_decompose(passenger["passengers"], period=12)

plt.figure(figsize=(10, 8))  # Set the figure size

# Original Time Series
plt.subplot(4, 1, 1)
plt.plot(passenger["passengers"])
plt.title("Original Time Series")

# Trend
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend)
plt.title("Trend")

# Seasonality
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal)
plt.title("Seasonality")

plt.subplot(4,1,4)
plt.plot(decomposition.resid)
plt.title("Noise")

plt.tight_layout()  # Adjust layout to prevent overlapping titles
plt.show()
```

```python
decomposition2 = sm.tsa.seasonal_decompose(passenger["passengers"], period=12, model="multiplicative")

plt.figure(figsize=(10,8))

plt.subplot(4,1 ,1)
plt.plot(passenger["passengers"])
plt.title("Original Series")

plt.subplot(4,1,2)
plt.plot(decomposition2.trend)
plt.title("Trend")

plt.subplot(4,1,3)
plt.plot(decomposition2.seasonal)
plt.title("Seasonality")

y = [1/8,2/8,3/8,4/8,5/8,6/8,7/8,8/8]
plt.subplot(4,1,4)
plt.scatter(decomposition2.resid,y)
plt.title("Noise")

```
